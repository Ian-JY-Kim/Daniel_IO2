```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from DGP_functions import *
import seaborn as sns
```

```python
# define the parameters
alpha = 1
beta = 1
delta = 1 
mu = 2
sigma = 1 
market_number = 250
```

```python
data_observable = DGP(alpha, beta, delta, mu, sigma, market_number)
data_observable_stat = data_observable.drop_duplicates(subset = ['market_id'])
```

```python
plt.hist(data_observable_stat['X_m'])
plt.title("market characteristics")
plt.show()
```

As expected market characteristics is normally distributed centered at 3.

```python
plt.hist(data_observable_stat['potential_entrant'])
plt.title("potential entrant")
plt.show()
```

Potential number of entrants is assumed to be equally distributed 2,3 and 4. And random generateing process generates expected outcome.

```python
plt.hist(data_observable_stat['N_star'])
plt.title("eqm number of entrants")
plt.show()
```

After solving for entry game, it turns out that most of the market have 1 or 2 entrants. One interesting point was there are markets with zero entrants.

```python
plt.scatter(data_observable_stat['potential_entrant'], data_observable_stat['N_star'])
plt.title("Num of Entrant across Market Structure")
plt.xlabel("Potential Number of Entrants")
plt.ylabel("Number of Entrants")
plt.show()
```

Every market structure includes outcomes where none of the firms enter the market.

```python
sns.histplot(data_entrant['Z_fm'], color = "blue", label = "entrants")
sns.histplot(data_exiter['Z_fm'], color = "red", label = "exiters")

plt.legend(title = "entry decision")
plt.title("cost shifter")
plt.show()
```

```python
sns.histplot(data_entrant['u_fm'], color = "blue", label = "entrants")
sns.histplot(data_exiter['u_fm'], color = "red", label = "exiters")

plt.legend(title = "entry decision")
plt.title("cost shock")
plt.show()
```
